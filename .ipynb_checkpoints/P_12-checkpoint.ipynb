## Outliers Filtering

All continuous variables but 'Recency' and 'NumStorePurchases' presented ouliers in their distributions.

```python
'Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 
'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumWebVisitsMonth', 'Age'
```

```python
trimmer = OutlierTrimmer(variables=['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 
                                    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumWebVisitsMonth', 'Age'], 
                         capping_method="iqr",
                         tail="both",
                         fold=1.5,
                         )
```

```python
trimmer.fit(data)
```

```python
print(data.shape)
data_ = trimmer.transform(data)
print(data_.shape)
```

We will check how our trimmer worked visualizing the variables new distributions:

```python
fig = plt.subplots(nrows=4,ncols=4,figsize = (11,7))

for i in range(len(continuous_fts)):
    plt.subplot(4,4,i+1)
    ax = sns.histplot(data_[continuous_fts[i]], color='dodgerblue', bins=20)

plt.suptitle('Continuous Variables distributions', fontsize=14)
plt.tight_layout()
```
